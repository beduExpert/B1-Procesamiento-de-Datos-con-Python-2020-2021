```python
missing_df = df.isna().sum(axis=0).reset_index()
missing_df.columns = ['column_name', 'missing_count']
missing_df['percentage_of_filled'] = (df.shape[0] - missing_df['missing_count']) / df.shape[0] * 100
missing_df = missing_df.sort_values('percentage_of_filled').reset_index(drop=True)

missing_df[missing_df['percentage_of_filled'] == 0]
```
